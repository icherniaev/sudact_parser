Find the number of cases for a given date

```python
browser = Chrome('/Users/ilyachernyaev/Documents/projects/chromedriver')
browser.implicitly_wait(10)
```

```python
links = []
for url in tqdm(urls):
    browser.get(url)
    for try_ in range(10):
        try:
            num_docs = soup.find('div', class_='prompting').text.split()[1]
            break
        except:
            sleep(1)
    # find the number of documents
    num_docs = int(soup.find('div', class_='prompting').text.split()[1])
    # define the number of pages
    if num_docs % 10 == 0:
        num_pages = int(num_docs / 10)
    else:
        num_pages = int((num_docs // 10) + 1)
    for page in tqdm(range(num_pages)):
        curr_url = url[:36] + f'{page}' + url[37:]
        browser.get(curr_url)
        for attempt_ in range(10):
            try:
                soup = BeautifulSoup(browser.page_source, 'html')
                results = soup.find('ul', class_='results').find_all('li')
                break
            except:
                sleep(1)
        for i in range(len(results)):
            link = 'https://sudact.ru' + results[i].find('a').get('href')
            links.append(link)
```
